Membaca Data URL

membaca data dari url import pandas as panda (library analisis data yang diperlukan untuk membersihkan data mentah ke dalam sebuah bentuk yang bisa untuk diolah)

```python
import pandas as pd

url='https://raw.githubusercontent.com/Ankit152/Fish-Market/main/Fish.csv'
c=pd.read_csv(url) 
c.head()
```

```python
a = c ['Width'] 
a
```

menentukan statistic deskriptif
import statistics as st (library yang digunakan untuk mencari perhitungan statistic seperti rata-rata, median dan modus)

```python
import statistics as st
```

MENCARI RATA-RATA

```python
st.mean(a)
```

MENCARI MEDIAN(Nilai Tengah)

```python
st.median(a)
```

MENCARI MODUS(Data yang sering muncul)

```python
st.mode(a)
```

```python
a.skew()
```
